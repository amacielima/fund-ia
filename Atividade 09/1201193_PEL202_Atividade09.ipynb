```python
import keras
import cv2
from matplotlib import pyplot as plt
from keras.applications.densenet import DenseNet121, DenseNet169, DenseNet201
from keras.applications import imagenet_utils
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
from keras.applications.densenet import preprocess_input
from keras.applications.densenet import decode_predictions
```

### Imagens

```python
# prepara imagens para os modelos
image_cmp = load_img('images/computer.jpg', target_size=(224, 224))
image_cmp = img_to_array(image_cmp)
image_cmp = image_cmp.reshape((1, image_cmp.shape[0], image_cmp.shape[1], image_cmp.shape[2]))
image_cmp = preprocess_input(image_cmp)

image_clk = load_img('images/clock.jpg', target_size=(224, 224))
image_clk = img_to_array(image_clk)
image_clk = image_clk.reshape((1, image_clk.shape[0], image_clk.shape[1], image_clk.shape[2]))
image_clk = preprocess_input(image_clk)

# plota as imagens
img_cmp = cv2.imread('images/computer.jpg', 1)
img_clk = cv2.imread('images/clock.jpg', 1)

plt.subplot(121)
plt.imshow(img_cmp)
plt.subplot(122)
plt.imshow(img_clk)
plt.show()
```

### DenseNet121

```python
densenet121_model = DenseNet121(include_top=True, weights='imagenet', input_tensor=None, input_shape=None, 
                                pooling=None, classes=1000)
```

```python
print('Predies das Imagens com o DenseNet121')
dn121_pred_cmp = densenet121_model.predict(image_cmp)
dn121_label_cmp = decode_predictions(dn121_pred_cmp)
print('\nNotebook:')
print(dn121_label_cmp[0][0])
print(dn121_label_cmp[0][1])
print(dn121_label_cmp[0][2])

dn121_pred_clk = densenet121_model.predict(image_clk)
dn121_label_clk = decode_predictions(dn121_pred_clk)
print('\nRelgio:')
print(dn121_label_clk[0][0])
print(dn121_label_clk[0][1])
print(dn121_label_clk[0][2])
```
